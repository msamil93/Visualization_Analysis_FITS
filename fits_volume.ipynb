```python
import numpy as np
import plotly.graph_objs as go
from astropy.io import fits

fits_file_path = 'C:/Users/ASUS/Downloads/MC-100_new.fits'
hdulist = fits.open(fits_file_path)
fits_data = hdulist[0].data

fits_data = np.nan_to_num(fits_data, nan=0.0, posinf=np.nanmax(fits_data), neginf=np.nanmin(fits_data))
opacity = np.clip(fits_data / np.max(fits_data), 0, 1)

distance_choice = input("Choose 'near' or 'far' for distance calculation: ").lower()

x, y, z = np.mgrid[:opacity.shape[0], :opacity.shape[1], :opacity.shape[2]]
x = (x / np.max(x)) * (-47.30 + 47.50) - 47.50
y = (y / np.max(y)) * (248.70 - 248.30) + 248.30

if distance_choice == 'near':
    z = (z / np.max(z)) * (0.07 - 0.01) + 0.01
elif distance_choice == 'far':
    z = (z / np.max(z)) * (15.64 - 15.56) + 15.56

# Hacim grseli oluturma
fig = go.Figure(data=go.Volume(
    x=x.flatten(),
    y=y.flatten(),
    z=z.flatten(),
    value=opacity.flatten(),
    isomin=0.1,
    isomax=0.8,
    opacity=0.1,
    surface_count=25
))

distance_label = "Near distance (kpc)" if distance_choice == 'near' else "Far distance (kpc)"
fig.update_layout(
    scene=dict(
        xaxis=dict(title="Right Ascension (RA)"),
        yaxis=dict(title="Declination (Dec)"),
        zaxis=dict(title=distance_label)
    )
)

fig.show()
```
