```python
import os
import cv2
import numpy as np
import plotly.graph_objs as go

folder_path = "C:/Users/ASUS/Downloads/MC-100_new_images"


image_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
n_images = len(image_files)
image_width, image_height = 53, 49 

data = []

for i, filename in enumerate(image_files):
    image_path = os.path.join(folder_path, filename)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    
    image = cv2.resize(image, (image_width, image_height))

   
    image_flattened = image.reshape(-1, 3) / 255.0
    
    
    y, x = np.meshgrid(np.linspace(248.40, 248.60, image_width),
                       np.linspace(-47.45, -47.30, image_height))
    z = np.full_like(x, i)  

   
    slice_data = go.Mesh3d(
        x=x.flatten(),
        y=y.flatten(),
        z=z.flatten(),
        vertexcolor=image_flattened,
        opacity=0.5,
    )
    data.append(slice_data)


fig = go.Figure(data=data)

fig.update_layout(
    scene=dict(
        aspectmode='cube',
        xaxis=dict(nticks=5, range=[-47.45, -47.30], title='X axis', showgrid=True, dtick=-0.05),
        yaxis=dict(nticks=5, range=[248.40, 248.60], title='Y axis', showgrid=True, dtick=0.05),
      
    ),
    scene_camera=dict(
        eye=dict(x=1.5, y=1.5, z=0.1)
    )
)

fig.show()
```
